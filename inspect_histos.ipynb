```python
def draw_one(x,y,yerr, label=None, by=0, byerr=0):
    w= y+by
    e= sqrt(yerr**2+byerr**2)
    plt.errorbar(x,w,yerr=e,capsize=0, label=label, fmt='-o', markersize=3)
    return w,e
```

```python
def draw_one_set(data, name):
    tags= ['sig','offp','bbl','bbf']
    labs= ['signal','OffPeak','B bkg lepton', 'B bkg fake']
    for t,l in zip(tags,labs):
        h= data[name+'_'+t]
        [x,y,e]= h[:,]
        draw_one(x,y,e, label=l)
```

```python
def draw_one_stack(data, name):
    tags= ['sig','offp','bbl','bbf']
    labs= ['signal','OffPeak','B bkg lepton', 'B bkg fake']
    for j,(t,l) in enumerate(zip(tags,labs)):
        h= data[name+'_'+t]
        [x,y,e]= h[:,]
        if ( j == 0 ):
            bw=0
            be=0
        bw,be= draw_one(x,y,e, label=l, by=bw, byerr=be)
```

```python
def draw_one_set_pair(data,name):
    fig= plt.figure(figsize=(12,4))
    ax= plt.subplot(121)
    draw_one_set(data, name)
    plt.legend(loc='upper right')
    plt.ylim(ymin=0)
    plt.xlabel('NN output')

    ax= plt.subplot(122)
    draw_one_stack(data, name)
    plt.legend(loc='upper right')
    plt.xlabel('NN output')
    return fig
```

```python
fig1= draw_one_set_pair(data,'ss-BBandOff-mm')
```

```python
def draw_stack_bars(data, name, tags, labels):
    colors= ['g','b','c','r','y']
    fig= plt.figure(figsize=(6,4))
    h = data[name+'_'+tags[0]]
    [x,w,e]= h[:,10:]
    width= x[1]-x[0]
    c= colors[0]
    left= x-0.5*width
    plt.bar(left, w, width, color=c, ec=c, label=labels[0])
    bm= w
    for j,(t,l) in enumerate(zip(tags[1:],labels[1:])):
        h = data[name+'_'+t]
        c= colors[j+1]
        [x,w,e]= h[:,10:]
        plt.bar(left, w, width, color=c, ec=c, label=l, bottom=bm)
        bm= bm+w
        
    return fig
```

```python
def draw_data_mc(data, signs='ss', flavors='ee'):
    tags= ['offp','bbl','bbf','sig']
    labs= ['OffPeak','B bkg lepton', 'B bkg fake','signal']
    fig= draw_stack_bars(data, signs+'-BBandOff-'+flavors, tags, labs)
    tdict={'ss':'same-sign', 'os':'opposite-sign'}
    title= tdict[signs]+ ' ' + flavors
    plt.title(title)
    h= data[signs+'-OnPeak-'+flavors+'_data']
    [x,w,e]= h[:,10:]
    plt.errorbar(x,w,e, fmt='o', color='k', label='Data', markersize=3)
    plt.legend(loc='upper center')
    plt.xlabel('NN output')
```

```python
draw_data_mc(data, 'ss', 'ee')
draw_data_mc(data, 'ss', 'em')
draw_data_mc(data, 'ss', 'me')
draw_data_mc(data, 'ss', 'mm')
```

```python
draw_data_mc(data, 'os', 'ee')
draw_data_mc(data, 'os', 'em')
draw_data_mc(data, 'os', 'me')
draw_data_mc(data, 'os', 'mm')
```
