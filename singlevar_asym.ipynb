```python
def rawhistos(darrs, labels, warrs=None, range=None, bins=100, title=None, xlabel=None):
    '''
    Return a list of histogram pdfs and common binning
    *darr* an array of arrays to make histograms
    *labels* a list of labels for arrays in *darr*
    *warrs* an array of weighting arrays
    '''
    hs= []
    bns=None
    if ( warrs== None):
        warrs= np.array([None]*len(darrs))
    for d,lb,ws in zip(darrs, labels, warrs):
        h,b,p = plt.hist(d, bins=bins, range=range, weights=ws, alpha=0.3, histtype='stepfilled', label=lb)
        hs.append(np.array(h, dtype='f4'))
        bns=b
    legend()
    if ( title ): plt.title(title)
    if ( xlabel ): plt.xlabel(xlabel)
    return hs, bns
```

```python
def plot_stack(x, ylist, widths, labels, cweights=None, dataidx=None, title=None, xlabel=None):
    '''
    *x* is center of bins
    *ylist* is a list of arrays each has the same length as *x*
    *widths* is the width of each bin (same length as *x*)
    *labels* is the label of each component
    *cweights* is the constant weight for each array in *ylist*
    *dataidx* is the index that identifies which array in *ylist* is the "data" which will not be stacked.
       Instead it is plotted as data points with error bars
    '''
    if cweights==None:
        cweights= np.array([1]*len(ylist))
    else:
        cweights= np.array(cweights)
    ydata=None
    ymcs=np.array(ylist)
    lbs= np.array(labels)
    if ( dataidx != None ):
        s= np.array([True]* len(ylist))
        s[dataidx]=False
        ymcs= ymcs[s]
        ydata= ylist[dataidx]
        ylbs= lbs[s]
        cweights= cweights[s]
    bottom=None
    colors=['b','g','r','c','m']
    for y,lb,co,cw in zip(ymcs,ylbs,colors[:len(ymcs)],cweights ):
        wy= y*cw
        plt.bar(x-0.5*widths, wy, width=widths, linewidth=0, bottom=bottom, label=lb, color=co, alpha=0.7)
        if bottom==None : bottom= wy
        else: bottom+= wy
    if title: plt.title(title)
    if xlabel: plt.xlabel(xlabel)
    if ydata!=None:
        plt.errorbar(x, ydata, sqrt(ydata), fmt='o', mec='k', mfc='0.5', label=labels[dataidx], alpha=0.7, ecolor='k')
    legend(numpoints=1)
```

```python
# Draw one-dim asymmetry
def hist_asym(x, ypos, yneg, title=None, xlabel=None, xlim=None):
    sel= ypos+yneg>0
    if ( xlim ):
        sel = sel&(x>=xlim[0])&(x<=xlim[1])
    asym= (ypos[sel]-yneg[sel])/(ypos[sel]+yneg[sel])
    plt.plot(x[sel],asym,'bo')
    if (xlim): plt.xlim(xlim)
    if (title):plt.title(title)
    if (xlabel):plt.xlabel(xlabel)
    return x[sel], asym
```
