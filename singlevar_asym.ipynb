```python
def rawhistos(darrs, labels, warrs=None, range=None, bins=100, title=None, xlabel=None):
    '''
    Return a list of histogram pdfs and common binning
    *darr* an array of arrays to make histograms
    *labels* a list of labels for arrays in *darr*
    *warrs* an array of weighting arrays
    '''
    hs= []
    bns=None
    if ( warrs== None):
        warrs= np.array([None]*len(darrs))
    for d,lb,ws in zip(darrs, labels, warrs):
        h,b,p = plt.hist(d, bins=bins, range=range, weights=ws, alpha=0.3, histtype='stepfilled', label=lb)
        hs.append(h)
        bns=b
    legend()
    if ( title ): plt.title(title)
    if ( xlabel ): plt.xlabel(xlabel)
    return hs, bns
```

```python
def plot_stack(x, ylist, widths, labels, cweights=None, dataidx=None, title=None, xlabel=None):
    '''
    *x* is center of bins
    *ylist* is a list of arrays each has the same length as *x*
    *widths* is the width of each bin (same length as *x*)
    *labels* is the label of each component
    *cweights* is the constant weight for each array in *ylist*
    *dataidx* is the index that identifies which array in *ylist* is the "data" which will not be stacked.
       Instead it is plotted as data points with error bars
    '''
    if cweights==None:
        cweights= np.array([1]*len(ylist))
    else:
        cweights= np.array(cweights)
    ydata=None
    ymcs=np.array(ylist)
    lbs= np.array(labels)
    if ( dataidx != None ):
        s= np.array([True]* len(ylist))
        s[dataidx]=False
        ymcs= ymcs[s]
        ydata= ylist[dataidx]
        ylbs= lbs[s]
        cweights= cweights[s]
    bottom=None
    colors=['b','g','r','c','m']
    for y,lb,co,cw in zip(ymcs,ylbs,colors[:len(ymcs)],cweights ):
        wy= y*cw
        plt.bar(x-0.5*widths, wy, width=widths, linewidth=0, bottom=bottom, label=lb, color=co, alpha=0.7)
        if bottom==None : bottom= wy
        else: bottom+= wy
    if title: plt.title(title)
    if xlabel: plt.xlabel(xlabel)
    if ydata!=None:
        plt.errorbar(x, ydata, sqrt(ydata), fmt='o', color='k', label=labels[dataidx], 
    legend(numpoints=1)
```

```python
# arrays of a variable from data sources and with cuts
# pcm
dpcm_ep = [ d.pcm[s] for d,s in zip(ds,poselecs) ]
dpcm_en = [ d.pcm[s] for d,s in zip(ds,negelecs) ]

# bfweights
allbfweights = [ np.ones(len(ds[0])), np.ones(len(ds[1])), ds[2].bfweight, ds[3].bfweight ]
dbfw_ep = [ w[s] for w,s in zip(allbfweights,poselecs) ]
dbfw_en = [ w[s] for w,s in zip(allbfweights,negelecs) ]
```

```python
# Unweighted pcm
fig= plt.figure(figsize=(14,10))
ax= plt.subplot(221)
heposs_raw, bns= rawhistos(dpcm_ep, labels, range=pcmrange, title= 'e-plus', xlabel='Pcm (GeV)')
ax= plt.subplot(222)
henegs_raw, bns= rawhistos(dpcm_en, labels, range=pcmrange, title= 'e-minus', xlabel='Pcm (GeV)')
# BF weighted pcm
ax= plt.subplot(223)
heposs_bfw, bns= rawhistos(dpcm_ep, labels, warrs= dbfw_ep, range=pcmrange, title= 'e-plus', xlabel='Pcm (GeV)')
ax= plt.subplot(224)
henegs_bfw, bns= rawhistos(dpcm_en, labels, warrs= dbfw_en, range=pcmrange, title= 'e-minus', xlabel='Pcm (GeV)')
```

```python
# Stack up to compare with onpeak data
x= mids(bns)
widths= np.diff(bns)
fig= figure(figsize=(14,10))
# raw
ax= plt.subplot(221)
plot_stack(x, heposs_raw, widths, labels, const_wgts, dataidx=0, title='e-plus', xlabel='Pcm (GeV)')
ax= plt.subplot(222)
plot_stack(x, henegs_raw, widths, labels, const_wgts, dataidx=0, title='e-minus', xlabel='Pcm (GeV)')
# bfweighted
ax= plt.subplot(223)
plot_stack(x, heposs_bfw, widths, labels, const_wgts, dataidx=0, title='e-plus, BF-weighted', xlabel='Pcm (GeV)')
ax= plt.subplot(224)
plot_stack(x, henegs_bfw, widths, labels, const_wgts, dataidx=0, title='e-minus, BF-weighted', xlabel='Pcm (GeV)')
```
