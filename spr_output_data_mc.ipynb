```python
def midpoints(a):
    return 0.5*( a[:-1]+a[1:] )
```

```python
def get_ratio(a, b, ea=None, eb=None):
    r = a/b
    er = None
    if ( ea!= None and eb!= None):
        er = r*r* sqrt( ea**2/a**2 + eb**2/b**2 )
    return r,er
        
```

```python
def my_histogram(a, weight):
    range=(0,1)
    nb=200
    h, edge = np.histogram(a, bins=nb, range=range, weights=weight)
    eh, edge = np.histogram(a, bins=nb, range=range, weights=weight**2)
    eh= sqrt(eh)
    return h, eh, edge
```

```python
# create histograms
# data
datacut = (abs(da.deltaT)<15)&(da.dtErr1<3)
h_da, h_err_da, edge_da = my_histogram(da.spr[datacut], da.weight[datacut])
# MC and offpeak
mcdtcut= (abs(mc.deltaT)<15)&(mc.dtErr1<3)
sigsel= ((mc.evttype2==0)|(mc.evttype2==1))&mcdtcut
contsel= (mc.evttype2==-1)&mcdtcut
bblsel = (mc.evttype2>=1)&(mc.evttype2<=6)&mcdtcut
bbfsel = (mc.evttype2==7)&mcdtcut
bbsel = bblsel|bbfsel
h_mc, h_err_mc, edge_mc = my_histogram(mc.spr[mcdtcut], mc.weight[mcdtcut])
h_sig, h_err_sig, edge = my_histogram(mc.spr[sigsel], mc.weight[sigsel])
h_cont, h_err_cont, edge = my_histogram(mc.spr[contsel], mc.weight[contsel])
h_bbl, h_err_bbl, edge = my_histogram(mc.spr[bblsel], mc.weight[bblsel])
h_bbf, h_err_bbf, edge = my_histogram(mc.spr[bbfsel], mc.weight[bbfsel])
h_bb, h_err_bb, edge = my_histogram(mc.spr[bbsel], mc.weight[bbsel])
```

```python
fig1 = P.figure(figsize=(12,9))
left, width = 0.1, 0.8
rect1 = [left, 0.3, width, 0.6]
rect2 = [left, 0.1, width, 0.2]
ax11= fig1.add_axes(rect1)
mids = midpoints(edge_da)
P.errorbar(mids, h_da, fmt='o', yerr=h_err_da, ms=3, color='blue', label='Data')
P.errorbar(mids, h_mc, fmt='-', yerr=h_err_mc, color='red', label='MC+OffPeak')
P.errorbar(mids, h_cont, fmt='-', color='green', label='OffPeak')
P.errorbar(mids, h_bbl, fmt='-', color='magenta', label='BB lepton')
P.errorbar(mids, h_bbf, fmt='-', color='cyan', label='BB fake')
#P.errorbar(mids, h_sig, fmt='-', color='black', label='Signal')
P.bar(mids,h_sig, edge[1:]-edge[:-1], color='b', alpha=0.5, linewidth=0, label='Signal')
leg = P.legend(loc='upper right')
P.xticks([])
ax11.grid(axis='y')

# Ratio
r, er = get_ratio(h_da, h_mc, h_err_da, h_err_mc)
ax12= fig1.add_axes(rect2)
P.errorbar(mids, r, fmt='o', yerr= er)
P.xlabel('NN output')
P.ylabel('Ratio')
P.ylim(0.8, 1.2)
P.yticks([0.9,1.0,1.1])
ax12.grid(axis='y')
P.xlabel('NN output')
```
