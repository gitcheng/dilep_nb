```python
import numpy as np
import matplotlib.pyplot as plt
from utilities import *
```

```python
histfile= 'ntuple/V4/sprvars-reduce/spr_pdg11/histograms/histograms.npz'
```

```python
def draw_stack_bars(histograms, labels, data=None ):
    if ( len(histograms) != len(labels) ):
        print 'Error: lengths of histograms and labels do not match.'
        return
    
    colors= ['g','b','c','r','y']
    lowbin=10
    bm= None
    for h,lab,c in zip(histograms,labels,colors[0:len(labels)] ):
        [x,w,e]= h[:,lowbin:]
        width= x[1]-x[0]
        left= x-0.5*width
        if bm == None:
            bm = np.zeros(len(x))
        plt.bar(left, w, width, color=c, ec=c, label=lab, bottom=bm)
        bm = bm+w
    if ( data != None ):
        [x,w,e] = data[:,lowbin:]
        plt.errorbar(x,w,e, fmt='o', color='k', label='Data', markersize=2)
```

```python
def draw_one_set_stack(histfile, release='R24', signs=['nn','pp']):
    fig= plt.gcf()
    k=1
    for fl in ['ee','em','me','mm']:
        plt.subplot(220+k)
        k+=1
        hdata= fetch_histogram(histfile, release=release, mode='OnPeak', flavors=[fl], signs= signs, type='data')
        hbkg=[]
        labels= ['Off-peak', 'B bkg fake', 'B bkg lepton', 'Signal']
        for bm,bt in zip(['OffPeak','BB','BB','BB'],['data','bbf','bbl','sig']):
            hbkg.append( fetch_histogram(histfile, release='R24', mode=bm, flavors=[fl], signs=signs, type=bt) )
        draw_stack_bars(hbkg, labels, data= hdata)
        plt.legend(loc='best', numpoints=1)
        plt.xlabel('NN output')
        plt.title(release+' '+fl)
```

```python
fig= plt.figure(figsize=(12,8))
draw_one_set_stack(histfile, release='R24', signs=['nn','pp'])
fig= plt.figure(figsize=(12,8))
draw_one_set_stack(histfile, release='R24', signs=['np','pn'])
```
