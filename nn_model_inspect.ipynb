```python
import numpy as np
import pylab as P
import matplotlib.pyplot as plt
def get_mids(h): return h[0,:]
def get_wgts(h): return h[1,:]
def get_errs(h): return h[2,:]
params = {'legend.fontsize': 15,
          'legend.linewidth': 2,
          'axes.labelsize': 15,
          'ytick.labelsize': 15,
          'xtick.labelsize': 15}
P.rcParams.update(params)
```

```python
def get_ratio(a, b, ea=None, eb=None):
    r = a/b
    er = None
    if ( ea!= None and eb!= None):
        er = r*r* sqrt( ea**2/a**2 + eb**2/b**2 )
    return r,er
```

```python
def draw_one_pair_with_ratio(x1,y1,yerr1,x2,y2,yerr2, label1=None, label2=None, ypm=0.5):
    fig= plt.gcf()
    left, width, bottom, top = 0.1, 0.8, 0.1, 0.1
    divide= 0.4
    rect1 = [left, divide, width, 1.0-top-divide]
    rect2 = [left, bottom, width, divide-bottom]
    ax11= fig.add_axes(rect1)
    if ( (x1==x2).any == False ):
        print 'Error: x axis are different'
        print x1
        print x2
        return
    plt.errorbar(x1,y1,yerr=yerr1,capsize=0, label=label1, fmt='+', markersize=3)
    plt.errorbar(x2,y2,yerr=yerr2,capsize=0, label=label2, fmt='o', markersize=3)
    plt.xticks([])
    plt.grid(axis='y')
    plt.legend(loc='best')
    plt.yscale('log')
    plt.ylim(ymin=1.0)
    # Draw ratio
    r, er = get_ratio(y1, y2, yerr1, yerr2)
    ax12= fig.add_axes(rect2)
    plt.errorbar(x1, r, yerr= er, fmt='o', color='red', markersize=3)
    plt.xlabel('NN output')
    plt.ylabel('Ratio')
    plt.ylim( (1-ypm, 1+ypm) )
    dy= 0.1
    while ( ypm/dy > 3 ):
        dy = dy*2.0
    while ( ypm/dy < 2 ):
        dy = dy/2.0
    tks=[1]
    for vv in np.arange(dy, +ypm, dy):
        tks.append(1+vv)
        tks.append(1-vv)
    plt.yticks(tks)
    plt.grid(axis='y')
    plt.xlabel('NN output')
```

```python
def draw_one_pair_simple(x1,y1,yerr1,x2,y2,yerr2, label1=None, label2=None):
    if ( (x1==x2).any == False ):
        print 'Error: x axis are different'
        print x1
        print x2
        return
    sum1= '%.0f' % y1.sum()
    sum2= '%.0f' % y2.sum()
    plt.errorbar(x1,y1,yerr=yerr1,capsize=0, label=label1+':'+sum1, fmt='+', markersize=3)
    plt.errorbar(x2,y2,yerr=yerr2,capsize=0, label=label2+':'+sum2, fmt='o', markersize=3)
    plt.xticks([])
    plt.grid(axis='y')
    plt.yscale('log')
    plt.ylim((1,1e5))
    plt.xlabel('NN output')
    plt.legend(loc='best', numpoints=1)
```

```python
def draw_one_pair_by_name(data,name1, name2, label1=None, label2=None):
    h1= data[name1]
    h2= data[name2]
    [x1,y1,e1]= h1[:,]
    [x2,y2,e2]= h2[:,]
    draw_one_pair_simple(x1,y1,e1,x2,y2,e2,label1=label1,label2=label2)
```

```python
def draw_one_pair_by_name_with_ratio(data,name1,name2,label1=None, label2=None, ypm=0.5):
    h1= data[name1]
    h2= data[name2]
    [x1,y1,e1]= h1[:,]
    [x2,y2,e2]= h2[:,]
    draw_one_pair_with_ratio(x1,y1,e1,x2,y2,e2,label1=label1,label2=label2,ypm=ypm)
```

```python
for ff in ['ee','em','me','mm']:
    fig2= plt.figure(figsize=(15,6))
    i=1
    for co,lab in zip(['sig','bbl','bbf','offp'], ['signal','BB lepton', 'BB fake', 'off-peak'] ):
        ax1=plt.subplot(240+i)
        plt.title(ff+' '+lab, fontsize='x-large' ) 
        name1= 'hist_'+ff+'_pn_'+co
        name2= 'hist_'+ff+'_np_'+co
        label1= r'$+-$'
        label2= r'$-+$'
        draw_one_pair_by_name(data,name1,name2,label1,label2)
        plt.subplot(244+i)
        name1= 'hist_'+ff+'_pp_'+co
        name2= 'hist_'+ff+'_nn_'+co
        label1= r'$++$'
        label2= r'$--$'
        draw_one_pair_by_name(data,name1,name2,label1,label2)
        i=i+1
```
