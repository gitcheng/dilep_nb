```python
import matplotlib.pyplot as plt
```

```python
def quadsum(x,y):
    return sqrt(x*x+y*y)
```

```python
def drawLabel(x,y, lab1, lab2):
    plt.text(x,y+0.15, lab1, fontsize= 14) 
    plt.text(x,y-0.15, lab2, fontsize= 14)
```
