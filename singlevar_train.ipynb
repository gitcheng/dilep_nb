```python
def plot_allvars(datas, dlabels, variables, ranges=None, yscales=None):
    '''
    *datas* : an array of numpy structured arrays
    *dlabels* : a list of labels (for legend) associated with *datas*
    *variables* : a list of strings of variable names
    *ranges* : a dictionary that maps a variable name to its plotting range
    *yscales* : a dictionary that maps a variable name to a string that is passed to plt.yscale()
    '''
    counter=0
    for var in variables:
        rm = counter%3
        counter= counter+1
        if ( rm == 0 ):
            plt.figure(figsize=(14,3))
        ax= plt.subplot(1,3,rm+1)
        for da,lb in zip(datas,dlabels):
            x= da[var]
            if ( ranges ):
                if ( var in ranges ):
                    r1= ranges[var][0]
                    r2= ranges[var][1]
                    x= da[var][(da[var]>r1)&(da[var]<r2)]
            h= plt.hist(x, bins=100, alpha=0.4, label=lb, histtype='stepfilled')
        if ( yscales ):
            if var in yscales:
                plt.yscale(yscales[var])
        plt.xlabel(var)
        plt.legend()
```

```python
def RF_train(data, v_train, ntrsamples, labels, **rfargs):
    '''
    Random Forest training. Make probability histogram plots and return 
    ( classifier, histogram of probabilities of being 0 for data d0 and d1, histogram binning)
    *data* : list of data arrays
    *v_train* list of training variable names
    *ntrsamples*: list of number of traning samples
    *labels* :  list of labels for the legend
    other arguments will be passed to RandomForestClassifier()
    '''
    # Prepare train/test data
    Xs = [ d[v_train] for d in data ]
    # randomize
    idxs= [ np.random.permutation(len(X)) for X in Xs ]
    # unify record types in training array and convert to regular array
    dtypes= Xs[0].dtype
    ntypes=[]
    for t in dtypes.names: ntypes.append((t, '<f4'))
    Xs = [ X[idx].astype(ntypes).view(('<f4', len(dtypes.names))) for X,idx in zip(Xs, idxs) ]
    # train data
    Xs_train= [ X[:n] for X,n in zip(Xs, ntrsamples) ]
    X_train = np.concatenate(Xs_train)
    # target
    y_train= np.array( [ np.array([i]*len(X)) for i,X in enumerate(Xs_train) ] ).ravel()
    # test data
    Xs_test= [ X[n:] for X,n in zip(Xs, ntrsamples) ]
    # Classifier
    clf = RandomForestClassifier(**rfargs)
    clf.fit(X_train, y_train)
    # Calculate output
    ys= [ clf.predict_proba(X)[:,1] for X in Xs_test ]
    # Plot histograms
    hists, bins, patches= plt.hist( ys, bins=100, histtype='stepfilled', alpha=0.5, label=labels, normed=True)
    legend()
    return clf, hists, bins
```

```python
fig1= figure(figsize=(16,8))
v_train1= ['nTrk','thrust', 'R2', 'emiss','cosThetaS','cosThetaT','Lp0','Lp2']
v_train2= ['pflow'+ str(i) for i in xrange(9) ]
data= [d_off, d_BB]
dlabs= ['OffPeak', r'$B\overline{B}$']

xp=[]
yp=[]
clflabels=[]

plt.subplot(231)
clf11, hists11, bins= RF_train(data, v_train1, [50000,50000], labels=dlabs, n_estimators=10, min_samples_leaf=200)
xp.append(hists11[0])
yp.append(hists11[1])
clflabels.append('v1-10-200')

plt.subplot(232)
clf12, hists12, bins= RF_train(data, v_train1, [50000,50000], labels=dlabs, n_estimators=20, min_samples_leaf=200)
xp.append(hists12[0])
yp.append(hists12[1])
clflabels.append('v1-20-200')

plt.subplot(233)
clf13, hists13, bins= RF_train(data, v_train1, [50000,50000], labels=dlabs, n_estimators=20, min_samples_leaf=100)
xp.append(hists13[0])
yp.append(hists13[1])
clflabels.append('v1-20-100')

plt.subplot(234)
clf21, hists21, bins= RF_train(data, v_train2, [50000,50000], labels=dlabs, n_estimators=10, min_samples_leaf=200)
xp.append(hists21[0])
yp.append(hists21[1])
clflabels.append('v2-10-200')

plt.subplot(235)
clf22, hists22, bins= RF_train(data, v_train2, [50000,50000], labels=dlabs, n_estimators=20, min_samples_leaf=200)
xp.append(hists22[0])
yp.append(hists22[1])
clflabels.append('v2-20-200')

plt.subplot(236)
clf23, hists23, bins= RF_train(data, v_train2, [50000,50000], labels=dlabs, n_estimators=20, min_samples_leaf=100)
xp.append(hists23[0])
yp.append(hists23[1])
clflabels.append('v2-20-100')
```
