```python
def plot_allvars(datas, dlabels, variables, ranges=None):
    '''
    *datas* : an array of numpy structured arrays
    *dlabels* : a list of labels (for legend) associated with *datas*
    *variables* : a list of strings of variable names
    *ranges* : a dictionary that maps a variable name to its plotting range
    '''
    counter=0
    for var in variables:
        rm = counter%3
        counter= counter+1
        if ( rm == 0 ):
            plt.figure(figsize=(14,4))
        ax= plt.subplot(1,3,rm+1)
        for da,lb in zip(datas,dlabels):
            x= da[var]
            if ( ranges ):
                if ( var in ranges ):
                    r1= ranges[var][0]
                    r2= ranges[var][1]
                    x= da[var][(da[var]>r1)&(da[var]<r2)]
            h= plt.hist(x, bins=100, alpha=0.4, label=lb, histtype='stepfilled')
        plt.xlabel(var)
        plt.legend()
```

```python
vars_to_plot= ['pcm', 'plab', 'costhlab', 'thrust', 'R2', 'emiss', 'nTrk']
vranges= {'emiss': (-5,10), 'thrust': (0.5,1)}
plot_allvars([d_BB, d_off], [r'$B\overline{B}$', 'Off peak'], vars_to_plot, vranges)
```
