```python
def add_histograms(hists):
    '''
    hists is a list or an array of histograms
    A histogram is an array of shape (3,n), where [0][:] contains mid points in x, [1][:] contains the
    content of the histogram, [2][:] contains the errors of each bin.
    '''
    # Check shapes
    shape = hists[0].shape
    if ( shape[0] != 3 or len(shape)!=2):
        print shape
        raise TypeError('The histogram has a wrong shape')
        
    mids = hists[0][0]
    hsum = hists[0][1]
    herr = hists[0][2] ** 2
    for h in hists[1:]:
        # Check shape
        if ( h.shape != shape ):
            print 'first shape = ', shape, '. This shape = ', h.shape
            raise TypeError('The histograms have a different shape')
        # Check two arrays of mid points are identical
        test = mids==h[0]
        if ( test.prod() != 1 ):
            print mids
            print h[0]
            raise TypeError('The two histograms mid points are different.')
        hsum = hsum + h[1]
        herr = herr + h[2]**2
            
    return np.array( [ mids, hsum, np.sqrt(herr) ] )
        
```
